# **데이터 수집 자동화 코드**
### 수집용 코드 실행 이전 패키지 로드 필수 및
### 모든 수집은 크롤링코드 기반 이므로, 페이지 오픈코드는 동일

```python
# 수집용 패키지

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.alert import Alert

import time
from openpyxl import *
```

```python
# 질병별의약품통계_의료기관주질병연령5세구간별원내원외 자동화코드
for i in [chr(i) for i in range(65,66)] : # 코드 알파벳 @@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
    wb.save(i+'.xlsx')
    for j in range(51): # 코드 번호 @@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@ (51) or (51:100)
        # 코드 검색
        driver.find_element(By.XPATH, value='//*[@id="searchWrd"]').click()
        # 이전 값을 지우기
        driver.find_element(By.XPATH, value='//*[@id="searchWrd1"]').clear()
        driver.find_element(By.XPATH, value='//*[@id="searchWrd1"]').send_keys(i+str(j).zfill(2))
        driver.find_element(By.XPATH, value='//*[@id="popSearchBtn"]').click()
        time.sleep(5)
        try:
            driver.find_element(By.XPATH, value='//*[@id="layerSearchTbody"]/tr/td[1]/a').click()
            # 검색
            driver.find_element(By.XPATH, value='//*[@id="searchBtn1"]').click()
            time.sleep(2)
            alert = Alert(driver)
            alert.accept()
            time.sleep(50)
            driver.find_element(By.TAG_NAME,value="body").send_keys(Keys.PAGE_DOWN)
            time.sleep(2)
            driver.find_element(By.XPATH, value='//*[@id="exlBtn"]').click()
            time.sleep(2)
            driver.find_element(By.TAG_NAME,value="body").send_keys(Keys.PAGE_UP)
            time.sleep(2)
            wb=load_workbook(i+'.xlsx')
            ws=wb.active
            try:
                wbd=load_workbook('질병별의약품통계_의료기관주질병연령5세구간별원내원외_202101_202401 ('+str(j-count)+').xlsx')
            except:
                wbd=load_workbook('질병별의약품통계_의료기관주질병연령5세구간별원내원외_202101_202401.xlsx')
            append_data_to_a(wbd, ws)
            wb.save(i+'.xlsx')
        except:
            driver.find_element(By.XPATH, value='//*[@id="layerSearch"]/div[2]/a').click()
            time.sleep(2)
            error = load_workbook('누락코드.xlsx')
            ews = error.active
            ews.append([i+str(j).zfill(2)])
            error.save('누락코드.xlsx')
            count+=1
```

```python
# 다빈도질병통계_질병연령별5세구간별 자동화코드

url = 'https://opendata.hira.or.kr/op/opc/olapHifrqSickInfoTab2.do'
driver.get(url)

# 순위 설정
driver.find_element(By.XPATH, value='//*[@id="rnk"]').click()
driver.find_element(By.XPATH, value='//*[@id="rnk"]/option[3]').click()
time.sleep(2)
count = 0
for i in range(1,18):
    wb = Workbook()
    wb.save('다빈도질병통계_질병연령별5세구간별_'+str(i*5)+'세미만.xlsx') # 시작연령을 이름으로 지정하여 합산용 엑셀파일 제작
    driver.find_element(By.XPATH, value='//*[@id="ageGubun"]').click()
    driver.find_element(By.XPATH, value=f'//*[@id="ageGubun"]/option[{str(i)}]').click()
    for j in range(1,3):
        for k in range(1,3):
            time.sleep(3)
            driver.find_element(By.XPATH, value=f'//*[@id="searchForm"]/div/div/span[{str(j)}]/div/span[{str(k)}]/label').click()
            driver.find_element(By.XPATH, value='//*[@id="searchBtn1"]').click()

            # 검색 중
            time.sleep(7)
            try:
                driver.find_element(By.XPATH, value='//*[@id="exlBtn"]').click()
            except:
                driver.find_element(By.TAG_NAME,value="body").send_keys(Keys.PAGE_DOWN)
                driver.find_element(By.XPATH, value='//*[@id="exlBtn"]').click()

            # 다운로드 중
            time.sleep(10)
            driver.find_element(By.TAG_NAME,value="body").send_keys(Keys.PAGE_UP)

            # 다운로드한 파일 추가
            wb=load_workbook('다빈도질병통계_질병연령별5세구간별_'+str(i*5)+'세미만.xlsx')
            ws=wb.active
            try:
                wbd=load_workbook(f'다빈도질병통계_질별연령별5세구간별_2023 ({str(count)}).xlsx')
            except:
                wbd=load_workbook('다빈도질병통계_질별연령별5세구간별_2023.xlsx')
            append_data_to_a(wbd, ws)
            wb.save('다빈도질병통계_질병연령별5세구간별_'+str(i*5)+'세미만.xlsx')

            count+=1
```

```python
# 질병별 약효분류군 사용실적 자동화코드

url = 'https://opendata.hira.or.kr/op/opc/olapMeftDivTab3.do#none'
driver.get(url)

# 필터링 생성용
driver.find_element(By.XPATH, value='//*[@id="searchWrd"]').click()
driver.find_element(By.XPATH, value='//*[@id="searchWrd1"]').send_keys('000')
driver.find_element(By.XPATH, value='//*[@id="searchBtn2"]').click()
time.sleep(2)
driver.find_element(By.XPATH, value='//*[@id="layerSearchTbody"]/tr/td[1]/a').click()

# 기간 정하기 (2021/01/01~2024/01/01)
driver.find_element(By.XPATH, value='//*[@id="sYm"]').click()
for i in range(3):
    driver.find_element(By.XPATH, value='/html/body/div[8]/div[1]/span[1]').click()
driver.find_element(By.XPATH, value='/html/body/div[8]/div[2]/div/div/span[1]').click()
driver.find_element(By.XPATH, value='//*[@id="eYm"]').click()
driver.find_element(By.XPATH, value='/html/body/div[9]/div[2]/div/div/span[1]').click()

count = 1
for i in range(110,400) :
    # 코드 검색
    driver.find_element(By.XPATH, value='//*[@id="searchWrd"]').click()
    # 이전 값을 지우기
    driver.find_element(By.XPATH, value='//*[@id="searchWrd1"]').clear()
    driver.find_element(By.XPATH, value='//*[@id="searchWrd1"]').send_keys(str(i))
    driver.find_element(By.XPATH, value='//*[@id="searchBtn2"]').click()
    time.sleep(2)
    try:
        if driver.find_element(By.XPATH, value='//*[@id="layerSearchTbody"]/tr/td[2]').text != '$':
            driver.find_element(By.XPATH, value='//*[@id="layerSearchTbody"]/tr/td[1]/a').click()
        else:
            driver.find_element(By.XPATH, value='//*[@id="layerSearch"]/div[2]/a').click()
            print(str(i)+'$')
            continue
        # 검색
        driver.find_element(By.XPATH, value='//*[@id="searchBtn1"]').click()
        time.sleep(20)
        try:
            driver.find_element(By.XPATH, value='//*[@id="exlBtn"]').click()
        except:
            print(str(i)+'엑셀')
            continue
        count+=1
    except:
        time.sleep(2)
        print(i)
        driver.find_element(By.XPATH, value='//*[@id="layerSearch"]/div[2]/a').click()
        print(str(i)+'코드')
        continue
```

```python
# 약효분류별 성분 자동화코드

url = 'https://www.health.kr/searchIngredient/KPICeffect.asp'
driver.get(url)

headers = {'User-Agent': 'Mozilla/5.0'}
html = requests.get(url, headers=headers)
soup = BeautifulSoup(html.text, 'html.parser')

# 예시 리스트
druglist = ['전신마취제','최면진정제']
drugdict = {}

for i in druglist:
    driver.find_element(By.XPATH, value='//*[@id="effect_searchword"]').send_keys(i)
    time.sleep(2)
    driver.find_element(By.XPATH, value='//*[@id="btn_search"]').click()
    time.sleep(2)
    driver.find_element(By.XPATH, value='//*[@id="sort"]/a[4]').click()

    soup = BeautifulSoup(driver.page_source, 'html.parser')

    drugdict[i] = []

    ul_id = soup.find_all('ul','depth04')
    ul_id1 = soup.find_all('ul','depth05')

    for j in ul_id:
        count = 2
        while True:
            try:
                drugdict[i].append(driver.find_element(By.XPATH, value=f'//*[@id="{j['id']}"]/li[{str(count)}]/div/a').text)
                time.sleep(1)
                count+=1
                continue
            except:
                time.sleep(1)
                break

    for j in ul_id1:
        count = 1
        while True:
            try:
                drugdict[i].append(driver.find_element(By.XPATH, value=f'//*[@id="{j['id']}"]/li[{str(count)}]/a').text)
                time.sleep(1)
                count+=1
                continue
            except:
                time.sleep(1)
                break

    driver.find_element(By.XPATH, value='//*[@id="effect_searchword"]').clear()

dd = pd.DataFrame(drugdict)

# 엑셀 파일로 저장
dd.to_excel("약효별성분.xlsx", index=False, engine='openpyxl')
```
