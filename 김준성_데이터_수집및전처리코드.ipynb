3. (다빈도) '입원외래구분' 열 데이터 구분

```python
data = pd.read_excel('다빈도질병통계_전체.xlsx')
current_label = ''
label_count = 0
for index, value in data['입원외래구분'].items():
    if value != current_label:
        current_label = value
        label_count += 1
    
    if current_label == '입원':
        if label_count == 1:
            data.at[index, '입원외래구분'] = '입원 (양)'
        elif label_count == 3:
            data.at[index, '입원외래구분'] = '입원 (한)'
    elif current_label == '외래':
        if label_count == 2:
            data.at[index, '입원외래구분'] = '외래 (양)'
        elif label_count == 4:
            data.at[index, '입원외래구분'] = '외래 (한)'
data.to_excel('다빈도질병통계_전체_업데이트.xlsx', index=False)
```

4. (약물소비) 열 이름 재지정 및 데이터 형식 문자열에서 정수로 전환

```python
def mdy0to1(mdy0):
    origin = mdy0.columns    
    mdy0.columns = mdy0.iloc[0]
    mdy1 = mdy0[1:]

    new_columns = []
    for i, col in enumerate(mdy1.columns):
        if col == "금액":
            new_columns.append(f"{origin[i-1]} 금액")
        elif col == "수량":
            new_columns.append(f"{origin[i]} 수량")
        else:
            new_columns.append(col)

    mdy1.columns = new_columns

    intlist = mdy1.columns.difference(['코드', '질병명', '원내원외구분', '연령구분'])
    mdy1[intlist] = mdy1[intlist].replace('-', '0').astype(int)
    return mdy1
```

5. (약물소비) 월 열 생성

```python
def mdydf(result_df):
    # 연도와 월을 추출할 수 있는 패턴
    year_month_pattern = r'(\d{4})년 (\d{2})월'

    # 새로운 데이터프레임 만들기
    result_df = pd.DataFrame(columns=['코드', '질병명', '원내원외구분', '연령구분', '연월', '수량', '금액'])

    # 열을 하나씩 처리
    for col in df.columns:
        # '수량'과 '금액' 열만 선택
        if '수량' in col or '금액' in col:
            # 연도, 월 정보 추출
            match = pd.Series(col).str.extract(year_month_pattern)
            if match.notna().all(axis=1).any():
                year, month = match[0][0], match[1][0]

                if '수량' in col:  # 수량 열인 경우
                    quantity_col = col
                    amount_col = quantity_col.replace('수량', '금액')  # '수량' 열에 해당하는 '금액' 열을 찾기
                    if amount_col in df.columns:  # '금액' 열이 존재하는지 확인
                        temp_df = df[['코드', '질병명', '원내원외구분', '연령구분']].copy()
                        temp_df['연월'] = f'{year}-{month}'
                        temp_df['수량'] = df[quantity_col]
                        temp_df['금액'] = df[amount_col]
                        result_df = pd.concat([result_df, temp_df[['코드', '질병명', '원내원외구분', '연령구분', '연월', '수량', '금액']]])

    # 결과 데이터프레임 확인
    result_df.reset_index(drop=True, inplace=True)
    result_df['월'] = result_df['연월'].str[-2:]

    df = result_df[['코드','원내원외구분','연령구분','수량','금액','월']]

    return df
```
